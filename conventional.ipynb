## PRE-PROCESSING

```python
raw = pd.read_csv('data/train.csv')
display(raw)
data = raw.copy()
save = data.copy()
```

```python
data = save.copy()

# HomePlanet
homeplanet_labels = {p: i for i, p in enumerate(data['HomePlanet'].unique())}
print(homeplanet_labels)

data['HomePlanet'] = data['HomePlanet'].replace(homeplanet_labels).astype(int)

display(data)
```

```python
save = data.copy()
```

```python
data = save.copy()

# CryoSleep
cryosleep_labels = {p: i for i, p in enumerate(data['CryoSleep'].unique())}
print(cryosleep_labels)

data['CryoSleep'] = data['CryoSleep'].replace(cryosleep_labels).astype(int)

display(data)
```

```python
save = data.copy()
```

```python
data = save.copy()

# Cabin -> Split to Cabin1, Cabin2, Cabin3
data['Cabin'].fillna('NaN/NaN/NaN', inplace=True)
cabin_list = ['Cabin1', 'Cabin2', 'Cabin3']
data = pd.concat([data, data['Cabin'].str.split('/', expand=True).rename(columns={i: c for i, c in enumerate(cabin_list)})], axis=1)
data = data.drop(columns=['Cabin'])

cabin_labels = {c: {k: i for i, k in enumerate(data[c].unique())} for c in cabin_list}
for c in cabin_list:
    data[c] = data[c].replace(cabin_labels[c]).astype(int)

display(data)
```

```python
save = data.copy()
```

```python
data = save.copy()

# Destination
destination_labels = {d: i for i, d in enumerate(data['Destination'].unique())}
print(destination_labels)

data['Destination'] = data['Destination'].replace(destination_labels).astype(int)

display(data)
```

```python
save = data.copy()
```

```python
data = save.copy()

# VIP
vip_labels = {v: i for i, v in enumerate(data['VIP'].unique())}
print(vip_labels)

data['VIP'] = data['VIP'].replace(vip_labels).astype(int)

display(data)
```

```python
save = data.copy()
```

```python
data = save.copy()
data.fillna(0, inplace=True) # Fill NaN with 0
```

```python
save = data.copy()
```

```python
data = save.copy()

# Transported
transported_labels = {t: i for i, t in enumerate(data['Transported'].unique())}
print(transported_labels)

labels = data['Transported'].replace(transported_labels).astype(int)
data.drop(columns=['Transported'], inplace=True)

display(data)
display(labels)
```

```python
save = data.copy()
```

```python
data = save.copy()
data.drop(columns=['PassengerId', 'Name'], inplace=True)

display(data)
```

## CLASSIFICATION

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
scores = cross_val_score(model, data.values, labels.values, cv=5)
print(scores.mean())
```
